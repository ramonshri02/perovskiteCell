#### Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Load Data

```python
data = pd.read_excel('Data/Perovskite_database_content_all_data.xlsx')
data.head()
```

```python
data.info()
```

A Total of 43025 data points with 410 features.

Keep only the Perovskite columns.

```python
starting_letter = 'P'
selected_columns = [col for col in data.columns if col.startswith(starting_letter)]
df_filtered = data[selected_columns]
```

```python
df_filtered.head()
```

```python
# Export the Perovskite only DataFrame
output_file_path = 'perovskiteOnlyDF.xlsx'  
df_filtered.to_excel(output_file_path, index=False)
```
