```python
#installing libraries
import numpy as np
import pandas as pd
from CBFV import composition
```

```python
# initial dataset
data = pd.read_excel('perovskiteCell/Data/Perovskite_database_content_all_data.xlsx')
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
#filtering only perovskite columns
col_filter = [col for col in data.columns if 'Perovskite' in col]
col_filter
```

```python
perov_data = data.filter(col_filter)
perov_data
```

```python
perov_data['Perovskite_additives_compounds'].isna().sum()
```

```python
# this is the main column we need to predict in the end, hence removing all rows having null values of it
perov_data['Perovskite_additives_concentrations'].isna().sum()
```

```python
perov_data_filtered = perov_data.dropna(subset = 'Perovskite_additives_concentrations')
perov_data_filtered
```

```python
#droping columns having more than 1000 null values out of total 5000 rows
col_to_drop = [col for col in perov_data_filtered.columns if perov_data_filtered[col].isna().sum()>1000]
col_to_drop
```

```python
#final dataframe: df, have to train model on it
df = perov_data_filtered.drop(col_to_drop, axis = 1)
df
```
